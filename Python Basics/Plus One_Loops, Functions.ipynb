# Plus One: 

Given a array of digit in which the array represents an integer, we want to add 'one' to the number. For example, the list \[1, 2, 3\] represents the number 123. When we add one, the result would be 124 and the output is \[1, 2, 4\]. 


The problem here is that, if the least significant digit is '9' we must add one to the digit on the left. If the second one is also '9', the same applies for the third one. Therefore, we need to take into account all the digits, and we should add an additional digit to the list when necessary. Here is one solution to the problem:

```python
digits = [1, 2, 3]
```

```python
check = True
i = 0

while check and i != -len(digits) :
    i -= 1
    digits[i] = (digits[i] + 1) % 10
    check = (digits[i] == 0)
    
    if digits[0] == 0:
        digits.insert(0, 1)

print(digits)  
```

Another solution is:

```python
digits = [1, 2, 3]
```

```python
def add_one(digits):
    for i in range(-1, -len(digits)-1, -1)  :
        digits[i] = (digits[i] + 1) % 10
        if digits[i] != 0:
            return(digits)
    digits.insert(0, 1)
    return(digits)

print(add_one(digits))
```
